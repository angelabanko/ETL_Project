```python
import pandas as pd
```

```python
netflix_data = "netflix_titles.csv"
disney_data = "datasets_disney_plus_shows.csv"
review_data = "datasets_tv_shows.csv"
```

```python
netflix_df = pd.read_csv(netflix_data)
disney_df = pd.read_csv(disney_data)
review_df = pd.read_csv(review_data)
```

```python
netflix_df.head()
```

```python
# There was a Netflix movie that had a different ID, but everything else was duplicated
# This code removes the duplicate entry
netflix_df.drop_duplicates(subset=['title', 'release_year', 'listed_in'], keep='last', inplace=True)
```

```python
# Export the cleaned df to a csv
netflix_df.to_csv('clean_netflix_data.csv', index=False)
```

```python
disney_df.head()
```

```python
# Drop rows that contain blank values in the title column
# This was problematic for our PostGRES db because 'title' was the primary key
disney_df.dropna(subset = ["title"], inplace=True)
```

```python
# Export the cleaned dataframe to a csv
disney_df.to_csv('clean_disney_data.csv', index=False)
```

```python
review_df.head()
```

```python
# Drop the unamed index column from the dataframe
review_df.drop(columns='Unnamed: 0', inplace=True)
```

```python
# Export the cleaned dataframe to a csv
review_df.to_csv('clean_review_data.csv', index=False)
```
